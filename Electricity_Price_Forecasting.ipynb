```python
# import libraries.
import numpy as np
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import pandas as pd
import math as mt
import random
```

```python
data = pd.read_csv('/content/drive/MyDrive/EV Scheduling/ANN_Data_2.csv')
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
load_data = scalar.fit_transform(data)
print(load_data)
```

```python
df = pd.DataFrame(load_data)
df
```

```python
a = df.iloc[0:69985,0:99]
b = df.iloc[69985:,0:99]
print(a.shape)
print(b.shape)
```

```python
x = a.iloc[:,0:98].values
y = a.iloc[:,98].values
x1 = b.iloc[:, 0:98].values
y1 = b.iloc[:, 98].values
print(x.shape)
print(y.shape)
```

```python
print(x)
print('--------------')
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
import keras  #Keras is the deep learning library that helps you to code Deep Neural Networks with fewer lines of code
#from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop,Adadelta,SGD,Adagrad,Adam
```

```python
model = Sequential()
model.add(Dense(78,activation='relu', input_shape=(98,)))
model.add(Dense(78,activation='relu'))
model.add(Dense(78,activation='relu'))
model.add(Dense(78,activation='relu'))
model.add(Dense(78,activation='relu'))
model.add(Dense(78,activation='relu'))
model.add(Dense(1,activation='sigmoid')) #output Layer
model.summary()
```

```python
model.compile(loss='MSE',
              optimizer=Adam(),
              metrics=['MSE'])
```

```python
# Write the Training input and output variables, size of the batch, number of epochs
history = model.fit(X_train,y_train,
                    batch_size=32,
                    epochs=10,verbose=1)
```

```python
# Write the testing input and output variables
score = model.evaluate(X_test, y_test, verbose=2)
print('Test loss:', score[0])
```

```python
model.save("my_model_relu_78_6.h5")
```

```python
min_val = data.min()
print(min_val)
max_val = data.max()
print(max_val)
```

```python
min_val = min_val[98]
max_val = max_val[98]
```

```python
from keras.models import load_model
from keras.metrics import MeanSquaredError

# Provide custom_objects to load_model for built-in metrics that might cause deserialization issues
model = load_model('/content/my_model_relu_78_6.h5', custom_objects={'MSE': MeanSquaredError()})

# List or array containing predicted values
predicted_values = model.predict(x1)

# List or array containing actual values
actual_values = y1

# Flatten the predicted values to convert them into a 1D array
predicted_values_flat = np.squeeze(predicted_values)
```

```python
print(predicted_values_flat.shape)
print(actual_values.shape)
```

```python
# Inverse normalization to get the original values
predicted_values_original = []
actual_values_original = []
for i in range(len(actual_values)):
  predicted_values_original.append(predicted_values_flat[i] * (max_val - min_val) + min_val)
  actual_values_original.append(actual_values[i] * (max_val - min_val) + min_val)
```

```python

# Creating a DataFrame to store the predicted and actual values
data = {'Actual': actual_values_original, 'Predicted': predicted_values_original}
df = pd.DataFrame(data)

# Specifying the file path in Google Drive to save the Excel file
file_path = '/content/predicted_actual_values_mape_unseen.xlsx'

# Writing the DataFrame to an Excel file in Google Drive
df.to_excel(file_path, index=False)

print(f"Predicted and actual values saved to {file_path}")
```
